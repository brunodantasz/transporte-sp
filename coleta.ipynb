```python
import geopandas as gpd
import duckdb
import os
```

```python
# Lê os 3 arquivos
metro = gpd.read_file("dados/MEOP25_RMSP_CEM.shp")
corredores = gpd.read_file("dados/ONCO25_RMSP_CEM.shp")
terminais = gpd.read_file("dados/ONTE25_RMSP_CEM.shp")

print(f"Estações de metrô: {len(metro)} registros")
print(f"Corredores de ônibus: {len(corredores)} registros")
print(f"Terminais de ônibus: {len(terminais)} registros")
```

```python
conn = duckdb.connect("dados/transporte_sp.duckdb")

# Converte geometria pra texto (WKT) pra salvar no DuckDB
metro["geometry"] = metro["geometry"].astype(str)
corredores["geometry"] = corredores["geometry"].astype(str)
terminais["geometry"] = terminais["geometry"].astype(str)

conn.execute("DROP TABLE IF EXISTS metro")
conn.execute("DROP TABLE IF EXISTS corredores")
conn.execute("DROP TABLE IF EXISTS terminais")

conn.execute("CREATE TABLE metro AS SELECT * FROM metro")
conn.execute("CREATE TABLE corredores AS SELECT * FROM corredores")
conn.execute("CREATE TABLE terminais AS SELECT * FROM terminais")

print("✅ Dados salvos no DuckDB!")
print("Tabelas:", conn.execute("SHOW TABLES").fetchall())
conn.close()
```
